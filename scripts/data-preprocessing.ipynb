```python
import pandas as pd
import pickle

from time import perf_counter
```

```python
data_path = "/home/mmarzec12/data/"
saving_limit = 10**7
```

```python
start = perf_counter()
with open(data_path+"np_games_info.pkl" , 'rb') as f:
    data = pickle.load(f)
    end = perf_counter()
    elapsed = end - start
    print("Loading the data took {} minutes.".format(round(elapsed/60, 2)))
```

Data structure:
* ndarray with ~100000 lists (each list represent one game)
* at each field we have a list containing following elements:
    * game id
    * game name
    * production year
    * a range of number of people recommended playing with, the approximate time duration for a gameplay, information of age restriction, average ratings from user
    * types
    * categories
    * dictionary with user ratings

```python
#games = pd.DataFrame(columns=["game_id", "game_name", "type", "category"])
games_d = {}
chunk_id = 1
start = perf_counter()
n_games = len(data)
stats = ['stat_fortrade', 'stat_own', 'stat_preordered', 'stat_prevowned',
                'stat_want', 'stat_wanttobuy', 'stat_wanttoplay', 'stat_wishlist']

i = 0
d = {}
for count, game_df in enumerate(data):

    game_id = game_df[0]
    to_append = {"game_id":game_id, "game_name":game_df[1],
                 "type":game_df[5], "category":game_df[6]}
    
    games_d[game_id] = to_append
    
    for user,user_d in game_df[7].items():
        d[i] = {"user_name":user, "game_id":game_id,
               "score":user_d["score"], "timestamp":user_d["time_stamp"]}
        
        if user_d["stat"]:
            for st in user_d["stat"]:
                d[i]["stat_"+st] = 1
        
        
        i += 1 
        if i%10**6==0:
            print(f"Additional 1 000 000 ratings has beed preprocessed.")
            
    if i > saving_limit:
        print("Converting dictionary to pandas dataframe...")
        cols = ["user_name", "game_id", "score", "timestamp"] + stats
        res = pd.DataFrame.from_dict(d, orient="index", columns=cols)
        res.to_csv(data_path+"ratings_chunk_{}.csv".format(chunk_id))
        end = perf_counter()
        elapsed = end - start
        print("Chunk number {} has been saved, that took {} minutes ({}/{} games added).".format(chunk_id,
                                                                                                 round(elapsed/60, 2),
                                                                                                 count+1, n_games))
        # updating some values
        #res = pd.DataFrame(columns=["stat", "score", "time_stamp", "game_id", "game_name", "user_name"])
        start = perf_counter()
        chunk_id += 1
        i = 0
        d={}

# saving last chunk (possibly less than "saving_limit") 
if i > 0:
    cols = ["user_name", "game_id", "score", "timestamp"] + stats
    res = pd.DataFrame.from_dict(d, orient="index", columns=cols)
    res.to_csv(data_path+"ratings_chunk_{}.csv".format(chunk_id))
    end = perf_counter()
    elapsed = end - start
    print("Chunk number {} has been saved, that took {} minutes (all games added).".format(chunk_id, round(elapsed/60, 2)))

# saving data about the games
print("Saving data about the games...")
games_df = pd.DataFrame.from_dict(games_d, orient="index")
games_df.to_csv(data_path+"games_data.csv")
print("...data about the games has been saved.")
```
